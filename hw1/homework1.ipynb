```python
import pandas as pd
import numpy as np
from apriori import Apriori
import pyfpgrowth

df = pd.DataFrame(pd.read_csv('201703_Taiwan_preprocessed.csv'))
```

## Task 1:
**Transaction Define:**
    
    1.(PM2.5,Humidity,Temperature)
    2.device_id = 74DA3895C538
    
**Discretization Method:**

    Devided by 10 / 20
    
**Algorithm:**

    Apriori Algorithm / FP Growth Tree


```python
df1=df[df['device_id']=='74DA3895C538'].loc[:,['PM2.5','Humidity','Temperature']]
df1=df1.round(-1)#devided by 10 using round
transactions = df1.values.tolist()
for i in transactions:
    i[0]='P'+str(i[0])
    i[1]='H'+str(i[1])
    i[2]='T'+str(i[2])
transactions
```

### 1-1.Apriori Algorithm / divided by 10 / min sup:500/4416 , min conf:0.7

```python
#https://github.com/luoyetx/Apriori
minsup = 500/4416 
minconf = 0.7
ap = Apriori(transactions, minsup, minconf)
# run algorithm
ap.run()
# print out frequent itemset
ap.print_frequent_itemset()
# print out rules
ap.print_rule()
```

### 1-2.FP Growth Tree / divided by 10 / min sup:500/4416 , min conf:0.7

```python
#https://pypi.python.org/pypi/pyfpgrowth
patterns = pyfpgrowth.find_frequent_patterns(transactions, 500)
#patterns
rules = pyfpgrowth.generate_association_rules(patterns, 0.7)
print('Frequent itemset:')
patterns
```

```python
print('Rules:')
rules
```

```python
transactions1_2 = df1.values.tolist()
for i in transactions1_2:
    i[0]='P'+str(int(i[0]/20)*20)#devided by 20
    i[1]='H'+str(int(i[1]/20)*20)
    i[2]='T'+str(int(i[2]/20)*20)
transactions1_2
```

### 1-4.FP Growth Tree / divided by 20 / min sup:500/4416 , min conf:0.7

```python
#https://pypi.python.org/pypi/pyfpgrowth
patterns = pyfpgrowth.find_frequent_patterns(transactions1_2, 500)
#patterns
rules = pyfpgrowth.generate_association_rules(patterns, 0.7)
print('Frequent itemset:')
patterns
```

## Task 3:
**Transaction Define:**
    
    1.(PM2.5,device_id)
    2.Date=2017-03-20
    3.Time=12:00~15:00
    4.前500筆資料
    
**Discretization Method:**

    PM2.5 devided by 10 in level A~I / 20 in level A~E 
    
**Algorithm:**

    Apriori Algorithm / FP Growth Tree

```python
df3=df[(df['Date']=='2017-03-20')&(df['Time']>='12:00:00')& (df['Time']<='15:00:00')].loc[:,['Time','device_id','PM2.5']]
df3=df3.iloc[:500,:]
tmp = df3.values.tolist()
dic={}
for i in tmp:
    if (i[0] in dic):
        if (i[2]<10):
            dic[i[0]][0].append(i[1])
        elif ((i[2]>=10) & (i[2]<20)):
            dic[i[0]][1].append(i[1])       
        elif ((i[2]>=20) & (i[2]<30)):
            dic[i[0]][2].append(i[1])
        elif ((i[2]>=30) & (i[2]<40)):
            dic[i[0]][3].append(i[1])
        elif ((i[2]>=40) & (i[2]<50)):
            dic[i[0]][4].append(i[1])
        elif ((i[2]>=50) & (i[2]<60)):
            dic[i[0]][5].append(i[1])
        elif ((i[2]>=60) & (i[2]<70)):
            dic[i[0]][6].append(i[1])
        elif ((i[2]>=70) & (i[2]<80)):
            dic[i[0]][7].append(i[1])
        else:
            dic[i[0]][8].append(i[1])
    else:
        dic[i[0]]=[['A'],['B'],['C'],['D'],['E'],['F'],['G'],['H'],['I']]
transactions3=[]
for i in dic.values():
    for j in i:
        transactions3.append(j)
transactions3
```

### 3-1.Apriori Algorithm /PM2.5 divided by 10 / min sup:19/171 , min conf:0.9

```python
minsup = 19/171
minconf = 0.9
ap = Apriori(transactions3, minsup, minconf)
# run algorithm
ap.run()
# print out frequent itemset
ap.print_frequent_itemset()
# print out rules
ap.print_rule()
```

### 3-2. FP Growth Tree/ PM2.5 divided by 10 / min sup:19/171 , min conf:0.9

```python
#https://pypi.python.org/pypi/pyfpgrowth
patterns = pyfpgrowth.find_frequent_patterns(transactions3, 19)
#patterns
rules = pyfpgrowth.generate_association_rules(patterns, 0.9)
print('Frequent itemset:')
patterns
```
