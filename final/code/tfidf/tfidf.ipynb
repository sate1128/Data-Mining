```python
books = []
for i in range(1,6):
    with open('src/00'+str(i)+'ssb.txt', 'r',encoding = "ISO-8859-1") as myfile:
        books.append(myfile.read().replace('\n', ''))
        
```

```python
import nltk
import math
import string

from nltk.corpus import stopwords
from collections import Counter
from nltk.stem.porter import*
```

```python
def get_tokens(text):
    lower = text.lower()
    remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)
    no_punctuation = lower.translate(remove_punctuation_map)
    tokens = nltk.word_tokenize(no_punctuation)

    return tokens
```

```python
def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))

    return stemmed
```

```python
def tf(word, count):
    return count[word] / sum(count.values())
def n_containing(word, count_list):
    return sum(1 for count in count_list if word in count)
def idf(word, count_list):
    return math.log(len(count_list)) / (1 + n_containing(word, count_list))
def tfidf(word, count, count_list):
    return tf(word, count) * idf(word, count_list)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def count_term(text):
    tokens = get_tokens(text)
    filtered = [w for w in tokens if not w in stopwords.words('english')]
    stemmer = PorterStemmer()
    stemmed = stem_tokens(filtered, stemmer)
    count = Counter(stemmed)
    return count

def main():
    keywords = {}
    countlist = []
    for book in books:
        countlist.append(count_term(book))
    for i, count in enumerate(countlist):
        print("Top words in document {}".format(i + 1))
        scores = {word: tfidf(word, count, countlist) for word in count}
        sorted_words = sorted(scores.items(), key = lambda x: x[1], reverse=True)
        for word, score in sorted_words[:30]:
            keywords[word]=score
            print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
        wordcloud = WordCloud().generate_from_frequencies(keywords)
        keywords={}
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.savefig('wordcloud_'+str(i+1)+'.png',dpi = 300)
        plt.show()

if __name__ == "__main__":
    main()
```
