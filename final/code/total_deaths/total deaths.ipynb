```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("deaths.csv")
df.sort_values(by = ['Allegiances'], ascending = True , inplace = True)
df
```

```python
labels = []
deaths = []

cur="Arryn"
cnt=0

for house in df['Allegiances']:
    if house == cur:
        cnt = cnt+1
    else:
        cur = house
        deaths.append(cnt)
        cnt = 1
    if house not in labels:
        labels.append(str(house))
        print(house)
deaths.append(cnt)
        
print(deaths)
```

```python
colors = ['red','blue','pink','brown','lime', 'cyan', 'grey','green', 'yellow', 'fuchsia', 'tan', 'navy']

fig, ax = plt.subplots()
#ax.pie(deaths,labels=labels,colors=colors,labeldistance = 1.1,autopct = '%3.1f%%',shadow = False, startangle = 90,pctdistance = 0.6)
ax.pie(deaths,labels=labels,colors=colors,labeldistance = 1.1,shadow = False, startangle = 90,pctdistance = 0.6)

plt.axis('equal')
handles, labels = ax.get_legend_handles_labels()
plt.legend(reversed(handles), reversed(labels),bbox_to_anchor=(1, 1), loc='best', borderaxespad=0.)
plt.savefig('total_deaths.png',dpi = 300, bbox_inches='tight')
plt.show()
```
