```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("pov_cnt.csv",encoding = "ISO-8859-1")

```

```python

df.sort_values(by = ['POV_cnt'], ascending = True , inplace = True)
df.reset_index(drop = True,inplace = True)
```

```python
def the_color(y_n):
    if y_n == 'Yes':
        return 'green'
    else :
        return 'red'
```

```python
plt.figure(figsize=(8, 8))
plt.barh(y = df.index, width = df['POV_cnt'],align="center", tick_label = df['Character']\
,color = df['Main'].apply(the_color))
```

```python
for x,y in zip(df.index , df['POV_cnt']):
    plt.text(y+1, x, '%d' % y, ha='center', va= 'center')
plt.text(40, 3, 'Main Character', ha='center', va= 'center', fontsize=14, color = 'green')
plt.text(40, 2, 'Other Character', ha='center', va= 'center', fontsize=14, color = 'red')
plt.xlabel('The times of POV', fontsize=16)
plt.ylabel('Character' , fontsize=18)
plt.savefig('POV_of_cht.png', dpi = 500, bbox_inches='tight')
plt.show()
```
